```python
# Suponiendo que df es tu DataFrame original
def classify_additional_info(row):
    # Plantilla para las columnas
    info = {'transmission': None, 'horsepower': None, 'year': None, 'mileage': None}

    try:
        
        # Iterar por los elementos separados
        for value in row:
            # value = value.strip()  # Limpiar espacios extra
            if 'km' in value:  # Si contiene 'km', es mileage
                info['mileage'] = value
            elif value.isdigit() and len(value) == 4:  # Si es un n√∫mero de 4 d√≠gitos, es year
                info['year'] = value
            elif 'caballos' in value:  # Si contiene 'caballos', es horsepower
                info['horsepower'] = value

            elif ('Gasolina' or 'Di√©sel' or 'El√©ctrico' or 'Otros') in value:
                info['fuel'] = value
            elif ('Manual' or 'Autom√°tico') in value:  # Cualquier otro valor, considerarlo transmission
                info['transmission'] = value

    except:
        pass
    
    return pd.Series(info)
```

```python
df_wallapop[['transmission', 'horsepower', 'year', 'mileage', 'fuel']] = df_wallapop['additional_info'].str.split('¬∑').apply(classify_additional_info)
```
