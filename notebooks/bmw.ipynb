```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
bmw = '../raw_data/bmw.csv'

df_bmw = pd.read_csv(bmw)
```

```python
df_bmw
```

### Número de unidades por modelo (BMW)

```python
df_bmw.groupby(df_bmw['model']).size()
```

```python
df_bmw.groupby(df_bmw['model']).describe()
```

```python
bmw_prices = df_bmw.groupby(('model'))['price']
```

```python
bmw_prices.describe()
```

```python
bmw_prices.head()
```

```python
plt.figure(figsize=(18, 6))  # (ancho, alto)
sn.boxplot(x='model', y='price', data=df_bmw)
```

```python
model_counts = df_bmw['model'].value_counts().reset_index()
```

```python
model_counts
```

```python
plt.figure(figsize=(18, 6))  # (ancho, alto)
sn.lineplot(x='model', y='count', data=model_counts, marker='o')
```
